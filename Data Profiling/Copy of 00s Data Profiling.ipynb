```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pydrive as pyd
import io
from pandas_profiling import ProfileReport
from google.colab import drive
from google.colab import files

uploaded = files.upload()
```

```python
df = pd.read_csv(io.BytesIO(uploaded['dataset-of-00s.csv']))
df.head()
```

```python
df.loc[df['loudness'] > 1.0, 'loudness'] = 1.0
```

```python
df["duration_mins"] = df["duration_ms"] / 60000
df.pop("duration_ms")
```

Let's make some Seaborn Pair Plots to see if we can spot some linear trends right off the bat

```python
#spotify = sns.load_dataset("df")
g = sns.pairplot(df, vars=["loudness", "energy", "danceability", "acousticness", "valence"])
g.map(sns.scatterplot)
plt.show()
```
