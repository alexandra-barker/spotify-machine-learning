```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
import pydrive as pyd
import io
from pandas_profiling import ProfileReport
from google.colab import drive
from google.colab import files
```

```python
sixties_dataset = files.upload()
seventies_dataset = files.upload()
eighties_dataset = files.upload()
nineties_dataset = files.upload()
aughts_dataset = files.upload()
tens_dataset = files.upload()
```

```python
sixties_df = pd.read_csv(io.BytesIO(sixties_dataset['dataset-of-60s.csv']))
seventies_df = pd.read_csv(io.BytesIO(seventies_dataset['dataset-of-70s.csv']))
eighties_df = pd.read_csv(io.BytesIO(eighties_dataset['dataset-of-80s.csv']))
nineties_df = pd.read_csv(io.BytesIO(nineties_dataset['dataset-of-90s.csv']))
aughts_df = pd.read_csv(io.BytesIO(aughts_dataset['dataset-of-00s.csv']))
tens_df = pd.read_csv(io.BytesIO(tens_dataset['dataset-of-10s.csv']))
```

```python
sixties_df = sixties_df.assign(decade = "60s")
seventies_df = seventies_df.assign(decade = "70s")
eighties_df = eighties_df.assign(decade = "80s")
nineties_df = nineties_df.assign(decade = "90s")
aughts_df = aughts_df.assign(decade = "00s")
tens_df = tens_df.assign(decade = "10s")
```

```python
alldecades_df = pd.concat([sixties_df, seventies_df, eighties_df, nineties_df, aughts_df, tens_df], axis = 0, ignore_index = True)
#alldecades_df
```

```python
alldecades_df.to_csv("dataset-alldecades.csv", index = False)
```
