Alternatively, we can check if there is any song duplicates:

```python
tracks_counts = df["track"].value_counts()
tracks_counts[tracks_counts > 1]
```

```python
df[df["track"]=="Breathe"]
```

"Breathe" is not duplicated since the songs are from different artists.

```python
df[df["track"]=="The Messenger"]
```

"The Messenger" by Turisas seems to be duplicated since both tracks have the same duration and share some other metrics. Investigating more, it seems that they are the same song released within two different albums.

Further investigation is neccessary to check for other duplicated songs in this dataset:

```python
list_repeated_songs = tracks_counts[tracks_counts > 1].index.to_list()
sorted(list_repeated_songs)
```

```python
repeated_df = df[df["track"].isin(list_repeated_songs)].sort_values(by="track")
repeated_df
```

```python
repeated_df.groupby(["artist", "track"])["artist"].count()[repeated_df.groupby(["artist", "track"])["artist"].count() > 1]
```

These should be all the duplicated songs.

```python
df[df["track"]=="Here Comes the Sun"]
```

Analyzing "Here Comes the Sun" by Yo-Yo Ma, both songs have the same metrics with a different uri. The two uris represent the same track in the same album.
